```python
from pathlib import Path
import os

def listar_archivos_proyecto(ruta_proyecto=None, excluir_carpetas=None):
    """
    Lista todos los archivos del proyecto excluyendo carpetas específicas.
    
    Args:
        ruta_proyecto (str): Ruta del proyecto (si es None, busca la raíz del proyecto)
        excluir_carpetas (list): Lista de nombres de carpetas a excluir
    
    Returns:
        list: Lista de rutas de archivos
    """
    if excluir_carpetas is None:
        excluir_carpetas = ['.env', '.venv', 'venv', '__pycache__', '.git', '.vscode', 'node_modules']
    
    # Si no se especifica ruta, buscar la raíz del proyecto
    if ruta_proyecto is None:
        # Intentar encontrar la raíz del proyecto (donde está .git o el notebook actual)
        ruta_actual = Path.cwd()
        
        # Subir hasta encontrar una carpeta que parezca raíz del proyecto
        while ruta_actual != ruta_actual.parent:
            # Indicadores de raíz de proyecto
            if (ruta_actual / '.git').exists() or \
               (ruta_actual / 'requirements.txt').exists() or \
               (ruta_actual / 'setup.py').exists() or \
               (ruta_actual / 'README.md').exists():
                ruta_proyecto = ruta_actual
                break
            ruta_actual = ruta_actual.parent
        
        # Si no encontramos indicadores, usar el directorio actual
        if ruta_proyecto is None:
            ruta_proyecto = Path.cwd()
    else:
        ruta_proyecto = Path(ruta_proyecto)
    
    ruta_proyecto = ruta_proyecto.resolve()
    print(f"Buscando archivos en: {ruta_proyecto}")
    print(f"Excluyendo carpetas: {excluir_carpetas}")
    print("-" * 60)
    
    archivos = []
    
    # Recorrer todo el árbol de directorios
    for ruta in ruta_proyecto.rglob('*'):
        # Verificar si la ruta está en alguna carpeta excluida
        if any(parte in excluir_carpetas for parte in ruta.parts):
            continue
        
        # Solo agregar archivos (no directorios)
        if ruta.is_file():
            # Obtener ruta relativa al proyecto
            try:
                ruta_relativa = ruta.relative_to(ruta_proyecto)
                archivos.append(str(ruta_relativa))
            except ValueError:
                # Si no se puede obtener ruta relativa, usar ruta absoluta
                archivos.append(str(ruta))
    
    return sorted(archivos)

# Ejecutar la búsqueda
archivos = listar_archivos_proyecto(excluir_carpetas=['.env', '.venv', '__pycache__', '.git'])
print(f"\nTotal de archivos encontrados: {len(archivos)}")

# Mostrar todos los archivos encontrados
print("\n--- Lista completa de archivos ---")
for i, archivo in enumerate(archivos, 1):
    print(f"{i:3d}. {archivo}")
```
