```python
# Celda 1 - Configuración inicial
import sys
from pathlib import Path
from pyspark.sql import functions as F

# Agregar raíz del proyecto
project_root = Path.cwd().parent  # si el notebook está en notebooks/
sys.path.append(str(project_root))

from src.data_loader import DataLoader

# Crear UNA sola instancia
data_loader = DataLoader()  # esto ya inicia Spark
spark = data_loader.spark

print("✅ Spark iniciado correctamente")
```

```python
# Celda 2 - Cargar datos
print("Cargando DivisionTerritorial...")
df_div = data_loader.Cargar_Division_Territorial()

print("Cargando Distritos...")
df_dist = data_loader.Cargar_Distritos()

# Hacer el JOIN (igual que lo tenías)


df_join = df_div.join(
    df_dist,
    df_div["CodigoAdministrativo"] == df_dist["codigo_administrativo"],
    "inner"
)

print(f"✅ Join completado. Total registros: {df_join.count():,}")
df_join.show(5)
```
